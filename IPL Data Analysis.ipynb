## cleaning the data

<b>In the above data, the players initially segregated based on the following parameters.
1. Capped - batters, bowlers, all-rounders, wicketkeepers.
2. Uncapped - batters, bowlers, all-rounders, wicketkeepers.

The uncapped players are the playing the ipl for the first time, or relavant data is not available therefore
we have use their domestic T20 performance and T20 international performance.

```python
#segregating data - capped batters
#we have kept the batting parameters to restrict the for batter specific tendencies.
```

```python
batters = df.loc[(df['Player_Type'] == 'Batter')]

batters_new = batters.loc[(batters['Capped'] == 1)]

capped_batters = batters_new[['Player Name', 
                              'Team',
                              'Nationality',
                              'Matches_Played',
                              'Runs',
                              'Average',
                              'Strike_Rate']]
```

```python
capped_batters.head()
```

```python
#segregating data - capped bowlers
#we have only kept the bowling parameters, since we will purely  judge the bowlers on their bowling statistics.
```

```python
bowlers = df.loc[(df['Player_Type'] == 'Bowler ')]

bowlers_new = bowlers.loc[(bowlers['Capped'] == 1)]

capped_bowlers = bowlers_new[['Player Name', 
                              'Team',
                              'Nationality',
                              'Matches_Played',
                              'Wickets',
                              'Bowling_average',
                              'Economy',
                              'Bowling_Strike_Rate']]
```

```python
capped_bowlers.head()
```

```python
#segregating data - capped keepers
#we have only keeping the relavant parameters for the keepers, since bowling figures are not needed for the keepers.
```

```python
keepers = df.loc[(df['Player_Type'] == 'Keeper')]

keepers_new = keepers.loc[(keepers['Capped'] == 1)]

capped_keepers = keepers_new[['Player Name', 
                              'Team',
                              'Nationality',
                              'Matches_Played',
                              'Runs',
                              'Average',
                              'Strike_Rate',
                              'Catches',
                              'Run_outs',
                              'Stumps']]
```

```python
capped_keepers.head()
```

```python
#segregating data - capped allrounders
#we have kept the bowling and batting figures for the allrounder to analyze the all round performance.
```

```python
allrounders = df.loc[(df['Player_Type'] == 'Allrounder')]

allrounders_new = allrounders.loc[(allrounders['Capped'] == 1)]

capped_allrounders = allrounders_new[['Player Name', 
                              'Team',
                              'Nationality',
                              'Matches_Played',
                              'Runs',
                              'Average',
                              'Strike_Rate',
                              'Wickets',
                              'Bowling_average',
                              'Economy',
                              'Bowling_Strike_Rate']]
```

```python
capped_allrounders.head()
```

```python
#Cleaning the data by filling the value 0 for null or NAN.
```

```python
capped_batters = capped_batters.fillna(0)
capped_bowlers = capped_bowlers.fillna(0)
capped_keepers = capped_keepers.fillna(0)
capped_allrounders = capped_allrounders.fillna(0)
```

```python
#Checking the null values in the data
```

```python
print(capped_batters.isna().sum())
print(capped_bowlers.isna().sum())
print(capped_keepers.isna().sum())
print(capped_allrounders.isna().sum())
```
